## Pricing - Measures of centrality 

Calculate the mean, median, and mode for the price of products.
What's the average price point of products listed? How does this compare with the most common price point (mode)?

```python
df['price'].describe().round(2)
```

```python
mean = df['price'].mean().round(2)
median = df['price'].median()
mode = df['price'].mode()[0]

mean, median, mode
```

```python
# Average: 89,24. Mode: 9,99. 
```

## Pricing - Measures of Dispersion
Determine the variance, standard deviation, range, and interquartile range for product price.
How varied are the product prices? Are there any indicators of a significant spread in prices?


```python
variance = round(df['price'].var(), 2)
std_dev = round(df['price'].std(), 2)
min_price = round(df['price'].min(), 2)
max_price = df['price'].max()
range_price = max_price - min_price
quantiles = df['price'].quantile([0.25, 0.5, 0.75])

variance, std_dev, min_price, max_price, range_price, quantiles
```

```python
# The maximum price is huge! So the range is also huge, but most prices are on the lower end. 
```

```python
quantiles2 = df['price'].quantile([0.8, 0.9, 0.95, 0.99]) # just curious
quantiles2
```

## Pricing - Visualizations 
Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. If its hard to read these diagrams, think why this is, and explain how it could be solved..
Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing.

```python
# Histogram 
sns.histplot(df['price'], bins=100, kde=True)
plt.xlabel("pPrice")
plt.ylabel("Frequency")
plt.title("Distribution of Prices")
plt.show()

# I've tried a bunch of different nos of bins, but it seems there is no way to make this legible in a regular histogram. 
# This is due to the extreme outliers on the high side of the prices. Maybe a boxplot?
```

```python
# Plotting a boxplot for the 'price' column with a light blue color
sns.boxplot(data = df['price'], color="lightblue")
plt.show()

# Also quite awful. Maybe create quantiles? 
```

```python
# Solution: change scale using a log scal 

# Log-transform the prices
log_prices = np.log1p(df['price'])

plt.figure(figsize=(10, 6))
sns.boxplot(x=log_prices, color="lightblue")

# Set custom ticks with original price values
ticks = [0, 2, 4, 6, 8, 10, 12]
plt.xticks(ticks, [int(np.expm1(tick)) for tick in ticks])  # expm1 reverses log1p
plt.xlabel('Price')
plt.title('Price Distribution (Log Scale)')
plt.show()
```

## Product Ratings - Measures of Centrality:

Calculate the mean, median, and mode for the rating of products.
How do customers generally rate products? Is there a common trend?


```python
mean = df['stars'].mean().round(2)
median = df['stars'].median()
mode = df['stars'].mode()[0]

mean, median, mode  

# That doesn't look great ... The most frequently rated score is 0 and over 50% of scores are also 0. Higher outliers bring up the mean.
```

```python
df.stars.value_counts()  # Just to check, here we see that indeed, 0 stars is by far the most frequent rating
```

## Product Ratings - Measures of Dispersion:

Determine the variance, standard deviation, and interquartile range for product rating.
Are the ratings consistent, or is there a wide variation in customer feedback?


```python
variance_stars = round(df['stars'].var(), 2)
std_dev_stars = round(df['stars'].std(), 2)
min_stars = round(df['stars'].min(), 2)
max_stars = df['stars'].max()
range_stars = max_stars - min_stars
quantiles_stars = df['stars'].quantile([0.25, 0.5, 0.75])

variance_stars, std_dev_stars, min_stars, max_stars, range_stars, quantiles_stars

# The ratings are heavily skewed to the right
```

## Product Ratings - Shape of the Distribution:

Calculate the skewness and kurtosis for the rating column.
Are the ratings normally distributed, or do they lean towards higher or lower values?


```python
skewness_stars = df['stars'].skew()
kurtosis_stars = df['stars'].kurtosis()

skewness_stars, kurtosis_stars

# This suggests low skew and flat distribuion. But the median, mode and mean do suggest skewness. 
```
